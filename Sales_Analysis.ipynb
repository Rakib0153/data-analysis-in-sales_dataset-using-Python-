# Data Visualization

```python
# Total Male and Female in the dataset
sns.set(style='dark')
plt.figure(figsize=(10,8)) #it size of Visualiz page  
ax=sns.countplot(x='Gender',data=sales_data,saturation=100,palette='Blues')#palette means dynamic color 
for bars in ax.containers: #its used for add labels in top of the bars
    ax.bar_label(bars)
plt.show()
```

```python
sales_data
```

```python
#Its Shows the Male and female in each Age_Group
plt.figure(figsize=(20,8))
ax=sns.countplot(x='Gender',hue='Age_Group',data=sales_data,palette='CMRmap',saturation=100)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#Visualized That which state ordered Most based on full dataset
top_state=sales_data.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False)#simple mean of Group BY fun used to split the data into groups based on some criteria. 
plt.figure(figsize=(25,8))
ax=sns.barplot(x='State',y='Orders',data=top_state,saturation=2500)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Find the top 3/4 oders from which state
top_order=sales_data.loc[(sales_data['Orders']>=3)]
top_state=top_order.groupby(['State'],as_index=False)['Orders'].count().sort_values(by='Orders',ascending=False)
plt.figure(figsize=(35,10))
ax=sns.barplot(x='State',y='Orders',data=top_state)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# Its shows the Maritial_Status 
plt.figure(figsize=(20,8))
ax=sns.countplot(x='Marital_Status',data=sales_data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#in Which States Highest Marital people are there  
top_maratial=sales_data.loc[(sales_data['Marital_Status']>=1)]
top_maratial_state=top_maratial.groupby(['State'],as_index=False)['Marital_Status'].count().sort_values(by='Marital_Status',ascending=False)
top_maratial_state
plt.figure(figsize=(35,20))
ax=sns.barplot(x='State',y='Marital_Status',data=top_maratial_state)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Marital_status based on Gender
top_maratial_mf=top_maratial.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].count().sort_values(by='Amount',ascending=False)
top_maratial_mf
plt.figure(figsize=(6,5))
ax=sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=top_maratial_mf)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Which professional people are present most
plt.figure(figsize=(25,10))
ax=sns.countplot(x='Occupation',data=sales_data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Occupation compare by male and female 
plt.figure(figsize=(25,10))
ax=sns.countplot(x='Occupation',hue='Gender',data=sales_data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Which Category of products are most popular
plt.figure(figsize=(35,15))
ax=sns.countplot(x='Product_Category',data=sales_data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# top 10 most sold products 
top_most_product=sales_data.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(20,8))
ax=sns.barplot(x='Product_ID',y='Orders',data=top_most_product)
for bars in ax.containers:
    ax.bar_label(bars)
```
